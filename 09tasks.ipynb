## Задание 1. Устойчивость по Ляпунову

Проблема устойчивости в общем случае неразрешима в том смысле, что нет одного алгоритма, применимого ко всем возможным ДУ.

Кратко, для уравнения 
$$
\dot{x} = f(t, x), \qquad x(t_0) = x_0
$$
решение $x^*(t)$ **устойчиво** по Ляпунову, если любое малое отклонение в начале остается малым в будущем, т.е.

$$
\forall\varepsilon>0\;\exists\delta(t_0,\varepsilon)>0:\ 
\|x_0-x^*(t_0)\|<\delta\Rightarrow\forall t\geq t_0\ 
\|x(t)-x^*(t)\|<\varepsilon.
$$

Решение называется **неустойчивым**, если оно не является устойчивым по Ляпунову. 

Решение $x^*(t)$ называется **асимптотически устойчивым**, если оно устойчиво по Ляпунову и является притягивающим, т.е. выполняется условие
$$
\displaystyle \lim_{t\to \infty }\|x(t)-x^*(t)\|=0
$$
для любого решения $x(t)$ с начальными данными $x_{0}$, для которых выполняется неравенство при некотором $\delta_{0}$
$$
\displaystyle ||x_{0}-x^*(t_{0})||<\delta_{0}.
$$

Существуют разновидности асимптотической устойчивости: <a href="https://ru.wikipedia.org/wiki/%D0%A3%D1%81%D1%82%D0%BE%D0%B9%D1%87%D0%B8%D0%B2%D0%BE%D1%81%D1%82%D1%8C_(%D0%B4%D0%B8%D0%BD%D0%B0%D0%BC%D0%B8%D1%87%D0%B5%D1%81%D0%BA%D0%B8%D0%B5_%D1%81%D0%B8%D1%81%D1%82%D0%B5%D0%BC%D1%8B)#%D0%90%D1%81%D0%B8%D0%BC%D0%BF%D1%82%D0%BE%D1%82%D0%B8%D1%87%D0%B5%D1%81%D0%BA%D0%B0%D1%8F_%D1%83%D1%81%D1%82%D0%BE%D0%B9%D1%87%D0%B8%D0%B2%D0%BE%D1%81%D1%82%D1%8C">равномерно асимптотически устойчивое, асимптотически устойчивое в целом, эквиасимптотически устойчивое, равномерно асимптотически устойчивым в целом</a>.

*Важно отметить, что в теории ОДУ вводятся также понятия устойчивости по Лагранжу, Пуассону, Четаеву, а также понятие структурной устойчивости.*

Для автономных уравнений исследование на устойчивость по Ляпунову имеет смысл в первую очередь для положений равновесия.

Сначала рассмотрим классические подходы к анализу устойчивости по Ляпунову, в котором рассматривается аналитическое решение.

---

**Пример 1.1:** Исходя из определения устойчивости по Ляпунову, исследовать на устойчивость решение уравнения $\dot{x} = 1$.

1. Общее решение:
$$
x'(t) = 1 \quad\Rightarrow\quad x(t) = t + C, \quad C \in \mathbb{R}.
$$
где $C = x(0)$ - начальное условие.

2. Смотрим отклонение от фиксированного решения:
Пусть $x^*(t)$ - частное решение с $C = C_0$:
$$
x^*(t) = t + C_0.
$$
Рассмотрим решение $x(t)$ с начальным отклонением $\delta_0$:
$$
x(t) = t + (C_0 + \delta_0).
$$
Тогда разность решений:
$$
\Delta(t) = x(t) - x^*(t) = \delta_0 = \text{const}.
$$

3. Проверка устойчивости по Ляпунову. Требуется проверить условие:
$$
\forall \varepsilon > 0 \; \exists \delta > 0: \; |x(0) - x^*(0)| < \delta \;\Rightarrow\; \forall t \geq 0 \; |x(t) - x^*(t)| < \varepsilon.
$$
Из выражения для разности:
$$
|x(t) - x^*(t)| = |\delta_0|.
$$
Выбираем $\delta = \varepsilon$. Тогда если $\|x(0)-x^*(0)\| = |\delta_0| < \delta$, то для всех $t \geq 0$:
$$
|x(t) - x^*(t)| = |\delta_0| < \varepsilon.
$$
Условие выполнено.

4. Проверка асимптотической устойчивости. Требуется: $\lim_{t \to \infty} |x(t) - x^*(t)| = 0$, но
$$
\lim_{t \to \infty} |x(t) - x^*(t)| = \lim_{t \to \infty} |\delta_0| = |\delta_0| \neq 0 \quad (\text{если } \delta_0 \neq 0).
$$
Следовательно, асимптотической устойчивости нет.

5. Геометрическая интерпретация: все решения - параллельные прямые $x = t + C$. Начальное отклонение сохраняется, но не растёт.

6. Итог: решение устойчиво по Ляпунову, но не асимптотически.

---

**Пример 1.2:** Исходя из определения устойчивости по Ляпунову, исследовать на устойчивость нулевое решение уравнения $\dot{x} = x$.

1. Общее решение:
$$
\dot{x} = x \quad\Rightarrow\quad \frac{dx}{x} = dt \quad\Rightarrow\quad \ln|x| = t + C
$$
$$
|x| = e^{t + C} = e^{C}e^t \quad\Rightarrow\quad x(t) = C e^t, \quad C \in \mathbb{R}
$$
где $C = x(0)$ — начальное условие.

2. Фиксируем исследуемое решение и вводим возмущение:
* Нулевое решение: $x^*(t) \equiv 0$ (соответствует $C = 0$)
* Возмущённое решение: $x(t) = \delta_0 e^t$, где $\delta_0 = x(0)$ — малое начальное отклонение
* Разность решений: $\Delta(t) = x(t) - x^*(t) = \delta_0 e^t$

3. Проверка устойчивости по Ляпунову. Требуется проверить условие:
$$
\forall \varepsilon > 0 \; \exists \delta > 0: \; |x(0)-x^*(0)| < \delta \;\Rightarrow\; \forall t \geq 0 \; |x(t)-x^*(t)| < \varepsilon
$$
Здесь $|x(0)-x^*(0)| = |\delta_0|$, а $|x(t)-x^*(t)| = |\delta_0| e^t$. Для любого $\delta > 0$ выберем $\delta_0 = \delta/2 \neq 0$. Существует момент $t_0 = \ln\left(\frac{2\varepsilon}{\delta}\right)$, такой что при $t > t_0$:
$$
|x(t)-x^*(t)| = \frac{\delta}{2} e^t > \varepsilon
$$
Условие устойчивости **не выполняется**.

4. Проверка асимптотической устойчивости:
$$
\lim_{t \to \infty} |x(t)-x^*(t)| = \lim_{t \to \infty} |\delta_0| e^t = +\infty \quad (\text{при } \delta_0 \neq 0)
$$
Условие не выполнено - отклонение растёт неограниченно.

5. Геометрическая интерпретация:
- Уравнение автономное: $\dot{x} = f(x) = x$
- Положение равновесия: $f(x) = 0 \Rightarrow x^* = 0$
- Производная: $f'(0) = 1 > 0$
- Фазовый портрет: при $x > 0$ производная $\dot{x} > 0$ (движение вправо), при $x < 0$ производная $\dot{x} < 0$ (движение влево). Стрелки направлены от нуля.

6. Итог: нулевое решение уравнения неустойчиво по Ляпунову.

---

**Пример 1.3:** Исходя из определения устойчивости по Ляпунову, исследовать на устойчивость нулевое решение уравнения $\dot{x} = -x$.

1. Общее решение: уравнение линейное однородное. Интегрируем:
$$
\frac{dx}{dt} = -x \quad\Rightarrow\quad \frac{dx}{x} = -dt \quad\Rightarrow\quad \ln|x| = -t + C_1
$$
где $C_1$ — постоянная интегрирования. Потенцируем:
$$
|x| = e^{-t + C_1} = e^{C_1}e^{-t} \quad\Rightarrow\quad x(t) = \pm e^{C_1}e^{-t}
$$
Введём постоянную $C = \pm e^{C_1}$ (любого знака). Тогда:
$$
x(t) = C e^{-t}, \quad C \in \mathbb{R}.
$$
Если $x(t_0) = x_0$, то:
$$
x_0 = C e^{-t_0} \quad\Rightarrow\quad C = x_0 e^{t_0}
$$
Следовательно:
$$
x(t) = x_0 e^{-(t - t_0)}
$$

2. Фиксируем исследуемое решение и вводим возмущение:
- Нулевое решение: $x^*(t) \equiv 0$ (соответствует $x_0^* = 0$)
- Возмущённое решение с начальным условием $x(t_0) = x_0$, где $x_0$ — малое отклонение
- Разность решений:
$$
\Delta(t) = x(t) - x^*(t) = x_0 e^{-(t - t_0)}
$$

3. Проверка устойчивости по Ляпунову: Требуется проверить для произвольного начального момента $t_0$:
$$
\forall \varepsilon > 0 \; \exists \delta(t_0, \varepsilon) > 0: \; |x_0| < \delta \Rightarrow \forall t \geq t_0 \; |x_0 e^{-(t - t_0)}| < \varepsilon
$$
Заметим, что $|x_0 e^{-(t - t_0)}| = |x_0| e^{-(t - t_0)} \leq |x_0|$ для всех $t \geq t_0$, поскольку экспонента $e^{-(t - t_0)} \leq 1$.   
Выберем $\delta = \varepsilon$. Тогда если $|x_0| < \delta$, то для всех $t \geq t_0$:
$$
|x(t) - x^*(t)| = |x_0| e^{-(t - t_0)} \leq |x_0| < \varepsilon
$$
Условие устойчивости выполнено. Причём можно взять $\delta = \varepsilon$, т.е. $\delta$ не зависит от $t_0$ - устойчивость равномерная.

4. Проверка асимптотической устойчивости, требуется дополнительно: $\lim_{t \to \infty} |x(t) - x^*(t)| = 0$. Вычисляем предел:
$$
\lim_{t \to \infty} |x_0 e^{-(t - t_0)}| = |x_0| \cdot \lim_{t \to \infty} e^{-(t - t_0)} = 0
$$
для любого $x_0$ (включая ненулевое). Условие выполнено, т.е. решение асимптотически устойчиво.

5. Фазовый анализ:
- Уравнение автономное: $\dot{x} = f(x) = -x$
- Положение равновесия: $f(x) = 0 \Rightarrow x^* = 0$
- Производная в точке равновесия: $f'(0) = -1 < 0$
- Фазовый портрет: при $x > 0$ производная $\dot{x} = -x < 0$ движение влево (к нулю), при $x < 0$ производная $\dot{x} = -x > 0$ движение вправо (к нулю). Все фазовые стрелки направлены к нулю равновесие асимптотически устойчиво.

6. Итог: нулевое решение уравнения является асимптотически устойчивым по Ляпунову. Можно даже уточнить, что экспоненциально устойчивым, поскольку отклонение затухает по экспоненте $e^{-t}$.

---

**Пример 1.4:** Исходя из определения устойчивости по Ляпунову, исследовать на устойчивость решение уравнения $\dot{x} = -x + \sin{t}$.

1. Общее решение: линейное неоднородное первого порядка
$$
x(t) = C e^{-t} + \frac{1}{2}(\sin{t} - \cos{t})
$$
Если $x(t_0) = x_0$, то константа:
$$
C = \left(x_0 - \frac{1}{2}(\sin{t_0} - \cos{t_0})\right) e^{t_0}
$$

2. Выбираем частное решение для исследования:
Рассмотрим частное решение, соответствующее $C = 0$:
$$
x^*(t) = \frac{1}{2}(\sin{t} - \cos{t})
$$
Это периодическое решение с периодом $2\pi$. Возмущённое решение:
$$
x(t) = C e^{-t} + \frac{1}{2}(\sin{t} - \cos{t})
$$
Разность:
$$
\Delta(t) = x(t) - x^*(t) = C e^{-t}
$$
где $C = \left(x_0 - x^*(t_0)\right) e^{t_0}$.

3. Проверка устойчивости по Ляпунову. Требуется проверить:
$$
\forall \varepsilon > 0 \; \exists \delta > 0: \; |x_0 - x^*(t_0)| < \delta \Rightarrow \forall t \geq t_0 \; |x(t) - x^*(t)| < \varepsilon
$$
Но $|x(t) - x^*(t)| = |C| e^{-t} = |x_0 - x^*(t_0)| e^{-(t - t_0)}$
Заметим: $e^{-(t - t_0)} \leq 1$ для всех $t \geq t_0$
Выбираем $\delta = \varepsilon$. Тогда если $|x_0 - x^*(t_0)| < \delta$, то для всех $t \geq t_0$:
$$
|x(t) - x^*(t)| = |x_0 - x^*(t_0)| e^{-(t - t_0)} \leq |x_0 - x^*(t_0)| < \varepsilon
$$
Условие выполнено. Причём $\delta$ не зависит от $t_0$, т.е. так называемая равномерная устойчивость.

4. Проверка асимптотической устойчивости. Требуется: $\lim_{t \to \infty} |x(t) - x^*(t)| = 0$
Вычисляем:
$$
\lim_{t \to \infty} |x(t) - x^*(t)| = \lim_{t \to \infty} |x_0 - x^*(t_0)| e^{-(t - t_0)} = 0
$$
для любых начальных отклонений. Условие выполнено, т.е. асимптотическая устойчивость. Более того, сходимость экспоненциальная со скоростью $e^{-t}$.

5. Фазовый анализ (для соответствующей автономной системы):
Уравнение неавтономное ($t$ явно входит в правую часть), поэтому обычный фазовый портрет на плоскости $(x)$ не строится. Можно рассмотреть расширенную систему:
$$
\begin{cases}
\dot{x} = -x + \sin{\theta} \\
\dot{\theta} = 1
\end{cases}
$$
В этом случае имеем положение равновесия на торе: $x = \frac{1}{2}(\sin{\theta} - \cos{\theta})$, которое оказывается асимптотически устойчивым. 
Линеаризация относительно решения $x^*(t)$:
$$
\dot{\xi} = -\xi, \quad \text{где } \xi = x - x^*(t)
$$
что подтверждает экспоненциальную устойчивость.

6. Итог: решение $x^*(t) = \frac{1}{2}(\sin{t} - \cos{t})$ асимптотически устойчиво по Ляпунову. Более того все решения этого уравнения сходятся к $x^*(t)$ при $t \to \infty$, поскольку разность любых двух решений затухает как $e^{-t}$.


## Задание 2. Аналитические методы исследования устойчивости ОДУ первого порядка и их недостатки

| Метод | Суть | Недостатки / ограничения |
|-------|------|-------------------------|
| **Явное решение (аналитическая интеграция)** | Решаем $\dot{x} = f(t,x)$ явно и анализируем $x(t)$ при $t \to \infty$ | - Часто невозможно найти аналитическое решение для нелинейных и неавтономных ОДУ<br>- Решение может быть слишком громоздким для анализа устойчивости |
| **Линеаризация около равновесной точки** | $\eta = x - x^*$, $\dot{\eta} \approx f'(x^*) \eta$; знак $f'(x^*)$ определяет локальную устойчивость | - Работает только локально для малых отклонений<br>- Неприменимо, если $x^*$ не существует или не единственный<br>- Для негладких функций и сильно осциллирующих функций (например, $\sin(1/x)$) не применимо<br>- Зависящие от времени коэффициенты могут менять знак, что усложняет анализ |
| **Метод функции Ляпунова** | Строим $V(x) > 0$, $\dot{V}(x) \le 0$ → устойчивость | - Нет универсального способа построить $V(x)$<br>- Часто возможно только локально<br>- Для сложных нелинейных функций построение может быть аналитически невозможно |
| **Фазовый портрет / графический анализ** | Строим направления $\dot{x} = f(x)$ и точки пересечения с осью | - Работает только в 1D или 2D<br>- Для функций с бесконечно быстрыми осцилляциями ($\sin(1/x)$) построение невозможно<br>- Для неавтономных систем требуется добавлять ось времени, усложняя анализ |
| **Метод вариационного уравнения / отклонений** | $\eta = x - x^*$, $\dot{\eta} = f(t,x^*+\eta) - f(t,x^*) \approx a(t)\eta$, где $a(t) = \frac{\partial f}{\partial x}(t,x^*)$ | - Зависит от известного решения $x^*(t)$<br>- Для сильно нелинейных или осциллирующих функций $a(t)$ может быть неограничена<br>- Интеграл $\int a(t) dt$ часто невычислим аналитически |
| **Спектральные методы (Ляпуновские экспоненты, Floquet-мультипликаторы)** | Вычисление показателей роста малых возмущений вдоль траекторий | - Требуется явное решение или численное интегрирование вариационного уравнения<br>- Аналитический анализ часто невозможен |

---

Выводы

- Аналитические методы ограничены:
  - Не работают без явного решения $x^*(t)$ или равновесной точки;
  - Неприменимы для сильно нелинейных, осциллирующих или неограниченных функций;
  - Иногда дают только локальную, а не глобальную устойчивость.
- Для сложных ОДУ обычно применяют численные методы:
  - Интеграция соседних траекторий (эквивалент определению Ляпунова)
  - Численное вариационное уравнение
  - Численные функции Ляпунова

---

**Пример численного метода оценки устойчивости**

Основные шаги:
1. Берём исследуемое решение $x^*(t)$, находим его численно.
2. Запускаем метод Эйлера из двух близких начальных условий:
- $x_0^{(1)} = x^*(0) + \delta$ (возмущённая траектория);
- $x_0^{(2)} = x^*(0)$ (невозмущённая траектория, назовем опорной).
3. На каждом шаге вычисляем отклонение: $e_k = x^{(1)}_{k} - x^{(2)}_{k}$.
4. Строим дискретную функцию Ляпунова $V_k = |e_k|$.
5. Анализируем поведение $V_k$:
- $V_k \rightarrow 0$, асимптотическая устойчивость;
- $V_k$ ограничена, устойчивость;
- $V_k$ неограниченно растёт, неустойчивость.

Ограничена, т.е. остаётся в пределах конечного интервала, но не стремится к нулю. Отклонения могут не расти, но не обязательно затухают. Например, $\max(V_k) < C\delta$, где $C \in [2;5]$.

```python
import numpy as np
import matplotlib.pyplot as plt

def assess_stability(f, t_span, h, x0_ref, delta, steps):
    """
    Численная оценка устойчивости решения ОДУ методом Эйлера
    
    Parameters:
    f : function
        Правая часть ОДУ dx/dt = f(t, x)
    t_span : tuple
        Интервал времени (t_start, t_end)
    h : float
        Шаг метода Эйлера
    x0_ref : float
        Начальное условие для опорной траектории
    delta : float
        Малое начальное возмущение
    steps : int
        Количество шагов
    
    Returns:
    str : Результат анализа устойчивости
    np.array : Время t
    np.array : Опорная траектория x_ref
    np.array : Возмущенная траектория x_pert
    np.array : Отклонение error
    """
    # Время
    t = np.linspace(t_span[0], t_span[1], steps+1)
    
    # 1. Интегрируем опорную траекторию
    x_ref = np.zeros(steps+1)
    x_ref[0] = x0_ref
    
    for k in range(steps):
        x_ref[k+1] = x_ref[k] + h * f(t[k], x_ref[k])
    
    # 2. Интегрируем возмущённую траекторию
    x_pert = np.zeros(steps+1)
    x_pert[0] = x0_ref + delta
    
    for k in range(steps):
        x_pert[k+1] = x_pert[k] + h * f(t[k], x_pert[k])
    
    # 3. Вычисляем отклонение
    error = np.abs(x_pert - x_ref)
    
    # 4. Анализируем
    if error[-1] < 1e-6: # маленькое отклонение
        stability = "Асимптотически устойчиво"
    elif np.max(error) < 2*delta: # отклонение не более чем в 2 раза от начального
        stability = "Устойчиво" 
    elif error[-1] > 100*delta: # отклонение в 100+ раз
        stability = "Неустойчиво"
    else:
        stability = "Неопределённо"
    
    return stability, t, x_ref, x_pert, error

# Пример 2.1: dx/dt = 1
def f1(t, x):
    return 1.0

# Пример 2.2: dx/dt = x
def f2(t, x):
    return x

# Пример 2.3: dx/dt = -x
def f3(t, x):
    return -x

# Пример 2.4: dx/dt = -x + sin(t)
def f4(t, x):
    return -x + np.sin(t)

# параметры для всех примеров
h = 0.1
t_span = (0, 10)
steps = int((t_span[1] - t_span[0]) / h)
delta = 0.01

# Пример 2.1
print("\nПример 2.1: dx/dt = 1")
stability1, t1, x_ref1, x_pert1, error1 = assess_stability(f1, t_span, h, 0.0, delta, steps)
print(f"Результат: {stability1}")
print(f"Начальное отклонение: {error1[0]:.6f}")
print(f"Конечное отклонение: {error1[-1]:.6f}")

# Пример 2.2
print("\nПример 2.2: dx/dt = x")
stability2, t2, x_ref2, x_pert2, error2 = assess_stability(f2, t_span, h, 0.0, delta, steps)
print(f"Результат: {stability2}")
print(f"Начальное отклонение: {error2[0]:.6f}")
print(f"Конечное отклонение: {error2[-1]:.6f}")

# Пример 2.3
print("\nПример 2.3: dx/dt = -x")
stability3, t3, x_ref3, x_pert3, error3 = assess_stability(f3, t_span, h, 0.0, delta, steps)
print(f"Результат: {stability3}")
print(f"Начальное отклонение: {error3[0]:.6f}")
print(f"Конечное отклонение: {error3[-1]:.6f}")

# Пример 2.4
print("\nПример 2.4: dx/dt = -x + sin(t)")
stability4, t4, x_ref4, x_pert4, error4 = assess_stability(f4, t_span, h, 0.0, delta, steps)
print(f"Результат: {stability4}")
print(f"Начальное отклонение: {error4[0]:.6f}")
print(f"Конечное отклонение: {error4[-1]:.6f}")

# Визуализация результатов
fig, axes = plt.subplots(4, 2, figsize=(14, 16))

# Пример 2.1
axes[0, 0].plot(t1, x_ref1, 'b-', label='Опорная траектория')
axes[0, 0].plot(t1, x_pert1, 'r--', label='Возмущенная траектория')
axes[0, 0].set_title('Пример 2.1: dx/dt = 1')
axes[0, 0].set_xlabel('t')
axes[0, 0].set_ylabel('x(t)')
axes[0, 0].legend()
axes[0, 0].grid(True)

axes[0, 1].plot(t1, error1, 'g-', label='Отклонение')
axes[0, 1].axhline(y=delta, color='r', linestyle='--', label='Начальное отклонение')
axes[0, 1].set_title(f'Отклонение: {stability1}')
axes[0, 1].set_xlabel('t')
axes[0, 1].set_ylabel('|Δx(t)|')
axes[0, 1].legend()
axes[0, 1].grid(True)

# Пример 2.2
axes[1, 0].plot(t2, x_ref2, 'b-', label='Опорная траектория')
axes[1, 0].plot(t2, x_pert2, 'r--', label='Возмущенная траектория')
axes[1, 0].set_title('Пример 2.2: dx/dt = x')
axes[1, 0].set_xlabel('t')
axes[1, 0].set_ylabel('x(t)')
axes[1, 0].legend()
axes[1, 0].grid(True)

axes[1, 1].plot(t2, error2, 'g-', label='Отклонение')
axes[1, 1].axhline(y=delta, color='r', linestyle='--', label='Начальное отклонение')
axes[1, 1].set_title(f'Отклонение: {stability2}')
axes[1, 1].set_xlabel('t')
axes[1, 1].set_ylabel('|Δx(t)|')
axes[1, 1].legend()
axes[1, 1].grid(True)

# Пример 2.3
axes[2, 0].plot(t3, x_ref3, 'b-', label='Опорная траектория')
axes[2, 0].plot(t3, x_pert3, 'r--', label='Возмущенная траектория')
axes[2, 0].set_title('Пример 2.3: dx/dt = -x')
axes[2, 0].set_xlabel('t')
axes[2, 0].set_ylabel('x(t)')
axes[2, 0].legend()
axes[2, 0].grid(True)

axes[2, 1].plot(t3, error3, 'g-', label='Отклонение')
axes[2, 1].axhline(y=delta, color='r', linestyle='--', label='Начальное отклонение')
axes[2, 1].set_title(f'Отклонение: {stability3}')
axes[2, 1].set_xlabel('t')
axes[2, 1].set_ylabel('|Δx(t)|')
axes[2, 1].legend()
axes[2, 1].grid(True)

# Пример 2.4
axes[3, 0].plot(t4, x_ref4, 'b-', label='Опорная траектория')
axes[3, 0].plot(t4, x_pert4, 'r--', label='Возмущенная траектория')
axes[3, 0].set_title('Пример 2.4: dx/dt = -x + sin(t)')
axes[3, 0].set_xlabel('t')
axes[3, 0].set_ylabel('x(t)')
axes[3, 0].legend()
axes[3, 0].grid(True)

axes[3, 1].plot(t4, error4, 'g-', label='Отклонение')
axes[3, 1].axhline(y=delta, color='r', linestyle='--', label='Начальное отклонение')
axes[3, 1].set_title(f'Отклонение: {stability4}')
axes[3, 1].set_xlabel('t')
axes[3, 1].set_ylabel('|Δx(t)|')
axes[3, 1].legend()
axes[3, 1].grid(True)

plt.tight_layout()
plt.show()
```

Система уравнений движения выглядит следующим образом:

\begin{align}
\left\{
        \begin{array}{l}
        \displaystyle m_1 \ddot{x}_1 + (k_1 + k_2) x_1 - k_2 x_2 = 0, \\
        \displaystyle m_2 \ddot{x}_2 + k_2 (x_2 - x_1) = 0.
        \end{array}
        \right.
        \tag{3.1}
\end{align}

Рассмотрим поведение системы при внешнем гармоническом воздействии на первый груз: $F(t) = F_0 \cos{\omega t}$. Запишем систему (3.1) с учетом внешней силы:
\begin{align}
\left\{
        \begin{array}{l}
        \displaystyle m_1 \ddot{x}_1 + (k_1 + k_2) x_1 - k_2 x_2 = F_0 \cos{\omega t}, \\
        \displaystyle m_2 \ddot{x}_2 + k_2 (x_2 - x_1) = 0.
        \end{array}
        \right.
        \tag{3.2}
\end{align}

Зададим систему следующим образом:

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
from scipy.integrate import solve_ivp

def ode(t, X, m1, m2, k1, k2, F0, omega):
    x1, v1, x2, v2 = X
    dx1dt = v1
    dv1dt = ( -(k1 + k2) * x1 + k2 * x2 + F0 * np.cos(omega*t) ) / m1
    dx2dt = v2
    dv2dt = ( k2 * x1 - k2 * x2 ) / m2 
    return [dx1dt, dv1dt, dx2dt, dv2dt]
```

Выберем следующие параметры для решения системы. Начальные условия выберем в положении равновесия.

```python
m1 = 10.0 # масса, кг
k1 = 10.0 # жёсткость пружины, Н/м
m2 = 10.0 # масса, кг
k2 = 10.0 # жёсткость пружины, Н/м
F0 = 1.0  # амплитуда внешней силы
omega = 1.0 # частота

# начальные условия: x1_0, v1_0, x2_0, v2_0
X0 = [0.0, 0.0, 0.0, 0.0]
```

Найдем численное решение системы и построим графики смещения грузов.

```python
t_full = 50
t_span = (0, t_full)
t_eval = np.linspace(0, t_full, 2000)

sol = solve_ivp(ode, t_span, X0, args=(m1,m2,k1,k2,F0,omega,), method='DOP853',rtol=1e-8, atol=1e-8, t_eval=t_eval)

t = sol.t
x1 = sol.y[0]
x2 = sol.y[2]

plt.figure(figsize=(8,4))
plt.plot(t, x1, label="x1 (верхняя масса)")
plt.plot(t, x2, label="x2 (нижняя масса)")
plt.grid(True)
plt.xlabel("t")
plt.ylabel("смещение")
plt.legend()
plt.tight_layout()
plt.show()
```

Несмотря на гармоническое воздействие на груз $x_1$, начальное смещение груза $x_2$ полностью компенсирует колебания $x_1$. В этом случае второй груз $x_2$ выступает в роли пассивного демпфера ([passive tuned mass damper](https://en.wikipedia.org/wiki/Tuned_mass_damper)) — устройства, состоящего из подвижной массы и упругого элемента (пружины). Оно предназначено для снижения амплитуды механических колебаний конструкции за счёт поглощения энергии колебаний и, в данном случае, при указанных начальных условиях свело амплитуду колебаний $x_1$ к нулю.

Однако классический пример работы динамического гасителя колебаний основан не на специальном выборе начальных условий, а на резонансном взаимодействии масс в установившемся режиме вынужденных колебаний. Динамический гаситель работает именно в установившемся режиме, создавая силу, противоположную по фазе внешней силе на частоте настройки.

---

Рассмотрим уравнения движения в матричном виде с внешней силой $F(t) = F_0 \sin{\omega t}$:

$$
\begin{bmatrix}
m_1 & 0\\
0 & m_2
\end{bmatrix}
\begin{Bmatrix}
\ddot{x}_1\\
\ddot{x}_2
\end{Bmatrix}
+
\begin{bmatrix}
k_1 + k_2 & -k_2\\
-k_2 & k_2
\end{bmatrix}
\begin{Bmatrix}
x_1\\
x_2
\end{Bmatrix}
=
\begin{Bmatrix}
F_0 \sin\omega t\\
0
\end{Bmatrix}
$$

Для анализа установившихся гармонических колебаний применим метод комплексных амплитуд. Предположим, что решения имеют вид $x_j(t) = X_j e^{i\omega t}$, где $X_j$ комплексная амплитуда. Тогда $\ddot{x}_j = -\omega^2 X_j e^{i\omega t}$, и система дифференциальных уравнений сводится к алгебраической, которую запишем в матричном виде:
$$
([K] - \omega^2 [M])\{X\} = \{F\}
$$

Выразим:

$$
\begin{bmatrix}
k_1 + k_2 - \omega^2 m_1 & -k_2\\
-k_2 & k_2 - \omega^2 m_2
\end{bmatrix}
\begin{Bmatrix}
X_1\\
X_2
\end{Bmatrix}
=
\begin{Bmatrix}
F_0\\
0
\end{Bmatrix}
$$

следовательно:

$$
X_1 = \frac{F_0(k_2 - \omega^2 m_2)}{(k_1 + k_2 - \omega^2 m_1)(k_2 - \omega^2 m_2) - k_2^2}
$$

$$
X_2 = \frac{F_0 k_2}{(k_1 + k_2 - \omega^2 m_1)(k_2 - \omega^2 m_2) - k_2^2}
$$

Теперь определим

$$
\omega_{n1} = \sqrt{\frac{k_1}{m_1}}, \qquad
\omega_{n2} = \sqrt{\frac{k_2}{m_2}}
$$

Для системы без второй массы где только одна степень свободы ($x_2$ отсутствует) резонанс возникает, когда

$$
\omega = \sqrt{\frac{k_1}{m_1}} = \omega_{n1}
$$

Чтобы $X_1$ была равна нулю на этой частоте, необходимо выполнение условия

$$
k_2 - \omega^2 m_2 = 0 \Rightarrow \omega = \sqrt{\frac{k_2}{m_2}} = \omega_{n2}
$$

Таким образом, если $k_2$ и $m_2$ выбраны так, что

$$
\frac{k_2}{m_2} = \frac{k_1}{m_1}
$$

то $X_1$ будет равна нулю при $\omega = \omega_{n1}$. Поэтому $x_2$ выступает в роли динамического поглотителя, для которого

$$
\omega_{n1} = \omega_{n2}
$$

На этой частоте перемещение $X_2$ будет:

$$
X_2 = \frac{F_0 k_2}{k_2 (k_1 + k_2 - \omega^2 m_1) - k_2^2}
= \frac{F_0}{k_2}
$$

Как было показано выше, добавление вспомогательной массы приводит к нулевым отклонениям основной массы при $\omega = \omega_{n1} = \omega_{n2}$. Однако при этом возникают две резонансные частоты $\omega_{d1}$ и $\omega_{d2}$, на которых амплитуда $X_2(t)$ становится существенной. Таким образом, динамический поглотитель может быть полезен только тогда, когда частота возмущения постоянна.

Тогда возникает вопрос, каким образом правильно подобрать параметры? Простых подходов два:
1. На основе ограничений по пространству. Выбрать $k_2$ и $m_2$ так, чтобы
$$
\frac{k_2}{m_2} = \frac{k_1}{m_1}
$$
при этом амплитуда колебаний массы $m_2$ будет равна
$$
X_2 = -\frac{F_0}{k_2}
$$

2. На основе требуемого разнесения собственных частот системы. Две новые собственные частоты системы могут быть получены приравниванием определителя к нулю:
$$
\left|
\begin{matrix}
k_1 + k_2 - \omega^2 m_1 & -k_2\\
-k_2 & k_2 - \omega^2 m_2
\end{matrix}
\right| = 0
$$
что даёт:
$$
(k_1 + k_2 - \omega^2 m_1)(k_2 - \omega^2 m_2) - k_2^2 = 0
$$

Это приводит к уравнению:
$$
\omega^4 - \omega^2\left( \frac{k_1}{m_1} + \frac{k_2}{m_1} + \frac{k_2}{m_2} \right)
+ \frac{k_1 k_2}{m_1 m_2} = 0
$$
которое можно представить в виде:
$$
\omega^4 - \omega^2(\omega_{n1}^2 + \mu \omega_{n1}^2 + \omega_{n2}^2) + \omega_{n1}^2 \omega_{n2}^2 = 0
$$
где $\mu = \frac{m_2}{m_1}$ — коэффициент отношения масс.

Корни этого уравнения $\omega_{d1}$ и $\omega_{d2}$ удовлетворяют соотношениям:
$$
\omega_{d1}^2 \cdot \omega_{d2}^2 = \omega_{n1}^2 \cdot \omega_{n2}^2
$$
$$
\omega_{d1}^2 + \omega_{d2}^2 = \omega_{n1}^2 + \omega_{n2}^2 + \mu \omega_{n2}^2
$$

Но для дополнительной массы $\omega_{n1} = \omega_{n2}$, следовательно:
$$
\omega_{d1}^2 \cdot \omega_{d2}^2 = \omega_{n1}^4
$$
$$
\omega_{d1}^2 + \omega_{d2}^2 = \omega_{n1}^2 (2 + \mu)
$$

При увеличении коэффициента массы $\mu$ собственные частоты $\omega_{d1}$ и $\omega_{d2}$ будут всё дальше расходиться. Причем, что $\omega_{d1}$ всегда ближе к $\omega_{n1}$, чем $\omega_{d2}$.

Следует отметить, что $\omega$ - круговая (угловая) частота, измеряется в радиан в секунду (рад/с). Обычная частота $f$ измеряется в Гц (герцах, колебаний в секунду) и связана с уловой частотой соотношением: $\omega$ = $2 \pi f$ (рад/с), $f = \omega/(2\pi)$ (Гц).


```python
import numpy as np
import matplotlib.pyplot as plt

# Параметры системы
m1 = 1.0
k1 = 1000.0
F0 = 10.0
mu = 0.1
m2 = mu * m1
k2 = (k1/m1) * m2

# Собственные частоты
omega_n1 = np.sqrt(k1/m1)
omega_n2 = np.sqrt(k2/m2)

# Частотный диапазон
omega_range = np.linspace(0.5*omega_n1, 1.8*omega_n1, 1000)

# Функции расчета
def X1_without_abs(omega):
    return F0 / np.abs(k1 - omega**2 * m1)

def X1_with_abs(omega, m1, m2, k1, k2):
    numerator = F0 * (k2 - omega**2 * m2)
    denominator = (k1 + k2 - omega**2 * m1) * (k2 - omega**2 * m2) - k2**2
    return np.abs(numerator / denominator)

def X2_with_abs(omega, m1, m2, k1, k2):
    numerator = -F0 * k2
    denominator = (k1 + k2 - omega**2 * m1) * (k2 - omega**2 * m2) - k2**2
    return np.abs(numerator / denominator)

# Расчет
X1_no = X1_without_abs(omega_range)
X1_yes = X1_with_abs(omega_range, m1, m2, k1, k2)
X2_yes = X2_with_abs(omega_range, m1, m2, k1, k2)

# Расчет новых собственных частот
coeff = [1, 0, -(omega_n1**2 + mu*omega_n1**2 + omega_n2**2), 0, omega_n1**2 * omega_n2**2]
roots = np.roots(coeff)
omega_d = np.sort([np.real(r) for r in roots if np.isreal(r) and r > 0])

# График 1: Сравнение с гасителем и без
plt.figure(figsize=(10, 6))
plt.plot(omega_range, X1_no, 'b-', linewidth=2, label='Без гасителя')
plt.plot(omega_range, X1_yes, 'r-', linewidth=2, label='С гасителем')
plt.axvline(x=omega_n1, color='k', linestyle='--', alpha=0.7, label=f'$\\omega_{{n1}} = {omega_n1:.1f}$ рад/с')
if len(omega_d) >= 2:
    plt.axvline(x=omega_d[0], color='g', linestyle=':', alpha=0.7, label=f'$\\omega_{{d1}} = {omega_d[0]:.1f}$ рад/с')
    plt.axvline(x=omega_d[1], color='g', linestyle=':', alpha=0.7, label=f'$\\omega_{{d2}} = {omega_d[1]:.1f}$ рад/с')
plt.xlabel('$\\omega$ [рад/с]', fontsize=12)
plt.ylabel('$|X_1|$ [м]', fontsize=12)
plt.title('Динамический гаситель: $X_1$ с гасителем и без', fontsize=14)
plt.grid(True, alpha=0.3)
plt.legend(fontsize=11)
plt.yscale('log')
plt.tight_layout()
plt.show()

# График 2: Амплитуды X1 и X2
plt.figure(figsize=(10, 6))
plt.plot(omega_range, X1_yes, 'r-', linewidth=2, label='$X_1$ (основная масса)')
plt.plot(omega_range, X2_yes, 'purple', linewidth=2, label='$X_2$ (гаситель)')
plt.axvline(x=omega_n1, color='k', linestyle='--', alpha=0.7, label=f'$\\omega_{{n1}} = \\omega_{{n2}}$')
zero_idx = np.argmin(np.abs(omega_range - omega_n1))
plt.plot(omega_n1, X1_yes[zero_idx], 'ro', markersize=8)
plt.xlabel('$\\omega$ [рад/с]', fontsize=12)
plt.ylabel('Амплитуда [м]', fontsize=12)
plt.title('Амплитуды основной массы и гасителя', fontsize=14)
plt.grid(True, alpha=0.3)
plt.legend(fontsize=11)
plt.yscale('log')
plt.tight_layout()
plt.show()
```

Следует отметить, что системы в air track, как правило, являются слабо связанными системами (weakly coupled system). Механическая система, в которой взаимодействие между подсистемами очень мало, т.е. коэффициенты жёсткости или демпфирования малы. В такой системе каждая масса колеблется почти как отдельный независимый осциллятор, а влияние одной части на другую слабое. 

Хорошо связанная система (strongly coupled system). Механическая система, в которой сила взаимодействия между подсистемами сопоставима с их собственными жёсткостями. Здесь колебания подсистем сильно влияют друг на друга, и система ведёт себя как единый объект с несколькими ярко выраженными собственными модами. Например, в таких системах можно увидеть две чётко разделённые резонансные частоты.

Уравнения движения при внешнем гармоническом воздействии на первый груз:
\begin{align}
\left\{
        \begin{array}{l}
        \displaystyle m_1 \ddot{x}_1 + (c_1 + c_2)\dot{x}_1 - c_2 \dot{x}_2 + (k_1 + k_2)x_1 - k_2 x_2 = F_0 \cos{\omega t}, \\
        \displaystyle m_2 \ddot{x}_2 + c_2 \dot{x}_2 - c_2 \dot{x}_1 + k_2 x_2 - k_2 x_1 = 0.
        \end{array}
        \right.
        \tag{3.3}
\end{align}

Зададим систему уравнений, время интегрирования и диапазон частот.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp

def ode(t, X, m1, m2, k1, k2, c1, c2, F0, omega):
    x1, v1, x2, v2 = X
    a1 = (F0*np.cos(omega*t) - (c1 + c2)*v1 + c2*v2 - (k1 + k2)*x1 + k2*x2) / m1
    a2 = (c2*v1 - c2*v2 + k2*x1 - k2*x2) / m2
    return [v1, a1, v2, a2]

t_full = 200
t_span = [0, t_full]  # время интегрирования
t_eval = np.linspace(0, t_full, 4000)
X0 = [0, 0, 0, 0]  # начальные условия: равновесие

frequencies = np.linspace(0.01, 0.5, 150)  # диапазон частот, Гц
omega_values = 2 * np.pi * frequencies
```

Зададим параметры слабо связанной системы:

```python
# Параметры системы (weakly coupled)
m1, m2 = 1.0, 0.5 # масса, кг
k1, k2 = 1.0, 0.25 # жёсткость пружины, Н/м
c1 = 0.5  # фиксированное демпфирование c1 (0.1 или 0.5), кг/с
F0 = 2.0  # амплитуда колебаний
c2_values = [0.01, 0.05, 0.1, 0.2, 0.5]
```

Визуализируем амплитудно-частотную характеристику (АЧХ) системы: зависимость амплитуды вынужденных колебаний от частоты внешнего периодического воздействия.

```python
colors = ['blue', 'red', 'green', 'purple', 'orange']

plt.figure(figsize=(12, 8))

# добавляем вертикальные линии на ожидаемых резонансных частотах (рубая оценка)
omega1_natural = np.sqrt(k1/m1)  # первая собственная частота
omega2_natural = np.sqrt(k2/m2)  # вторая собственная частота
print(omega1_natural/(2*np.pi)) # ожидаемой резонансной частоты 
print(omega2_natural/(2*np.pi)) # ожидаемой резонансной частоты

# проходим по разным значениям c2
for c2, color in zip(c2_values, colors):
    print(f"c2={c2:.3f}")
    amplitudes = []
    for i, omega in enumerate(omega_values):  # для каждой частоты
        # Интегрируем уравнения
        sol = solve_ivp(ode, t_span, X0, args=(m1, m2, k1, k2, c1, c2, F0, omega), method='DOP853', rtol=1e-8, atol=1e-8, t_eval=t_eval)
        
        # берем последнюю часть решения для установившегося режима
        steady_state_idx = int(0.8 * len(sol.t))  # последние 20% времени
        x1_steady = sol.y[0, steady_state_idx:]
        t_steady = sol.t[steady_state_idx:]
        
        # амплитуда как максимальное отклонение в установившемся режиме
        amplitude = np.max(np.abs(x1_steady))
        amplitudes.append(amplitude)
    # Построение графика для данного c2
    plt.plot(frequencies, amplitudes, color=color, linewidth=2, label=f'c2={c2:.3f}')


plt.axvline(omega1_natural/(2*np.pi), color='red', linestyle='--', alpha=0.5, label=f'ω₁={omega1_natural/(2*np.pi):.3f} Гц')
plt.axvline(omega2_natural/(2*np.pi), color='red', linestyle=':', alpha=0.5, label=f'ω₂={omega2_natural/(2*np.pi):.3f} Гц')

plt.xlabel('Частота (Гц)')
plt.ylabel('Амплитуда')
plt.grid(True, alpha=0.3)
plt.ylim(0, 6)
plt.xlim(0, 0.5)
plt.tight_layout()
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp

def ode(t, X, m1, m2, k1, k2, k3, c1, c2, F0, omega):
    x1, v1, x2, v2 = X

    a1 = (F0*np.cos(omega*t)
          - (c1 + c2)*v1 + c2*v2
          - (k1 + k2)*x1 + k2*x2) / m1

    a2 = (c2*v1 - c2*v2
          + k2*x1 - (k2 + k3)*x2) / m2

    return [v1, a1, v2, a2]

t_full = 200
t_eval = np.linspace(0, t_full, 5000)
t_span = (0, t_full)
X0 = [0, 0, 0, 0]

frequencies = np.linspace(0.01, 1.2, 180)
omega_values = 2*np.pi*frequencies
```

```python
def plot_frf(m1, m2, k1, k2, k3, c1, c2_list, F0, title):

    plt.figure(figsize=(13, 6))

    for c2 in c2_list:
        A1, A2 = [], []

        for omega in omega_values:
            sol = solve_ivp(
                ode, t_span, X0,
                args=(m1, m2, k1, k2, k3, c1, c2, F0, omega),
                method='DOP853',
                rtol=1e-8,
                atol=1e-8,
                t_eval=t_eval
            )

            ss = int(0.8*len(sol.t))
            x1 = sol.y[0, ss:]
            x2 = sol.y[2, ss:]

            A1.append(np.max(np.abs(x1)))
            A2.append(np.max(np.abs(x2)))

        plt.plot(frequencies, A1, label=f'x₁, c₂={c2}')
        plt.plot(frequencies, A2, '--', label=f'x₂, c₂={c2}')

    plt.xlabel('Частота, Гц')
    plt.ylabel('Амплитуда')
    plt.title(title)
    plt.grid(alpha=0.3)
    plt.legend(ncol=2)
    plt.tight_layout()
    plt.show()
```
