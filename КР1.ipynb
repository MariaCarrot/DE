## Задача 2

```python
from sympy import *
import numpy as np
import matplotlib.pyplot as plt
# Начальные условия
m = 3400 #кг
v0 = 18 #м/с
k_d = 10
k_l = 30
mu = 0.05
g = 9.8

t = symbols("t", real=True)
v = Function("v")
a = Derivative(v(t), t)
```

```python
# Уравнение ma=-(k_d-mu*k_l)*v^2 - mumg
ode = Eq(m * a, -(k_d-mu*k_l)*v(t)**2 - mu*m*g)
ode
```

```python
sol = dsolve(ode, v(t))
sol
```

```python
# v(0)=v0
C1 = symbols("C1")
C_val = Eq(sol.rhs.subs(t, 0), v0)
C_val
```

```python
C_sol = solve(C_val, C1)[0]
C_sol
```

```python
# Время до остановки
# v = 0
t_stop = solve(Eq(sol.rhs.subs(C1, C_sol), 0), t)[0]
t_stop
```

```python
# длина пробега ds/dv = v
s = integrate(sol.rhs.subs(C1, C_sol), (t, 0, t_stop))
s
```
