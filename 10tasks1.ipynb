**Пример 1.**  $yy''-{y'}^2=0$

Вводим $y'=z$, $y''=z'z$.

Подставляем в уравнение: $yz'z=z^2$.

1. Первый вариант, если $z = 0$, т.е. $y'=0$. Тогда $\boxed{y=C}$ является решением.

2. Второй вариант. Предполагаем, что $z \neq 0$ и разделим обе части на $z$ и разделим переменные.
$$
yz'=z \quad \Rightarrow y\dfrac{dz}{dy}=z \Rightarrow \dfrac{dz}{z} = \dfrac{dy}{y}
$$
Интегрируем
$$
\ln|z| = \ln|y| + C_1, \quad \Rightarrow \quad z = C_2 y
$$
Выполняем обратную замену и снова разделяем переменные
$$
z = C_2 y \quad \Rightarrow \quad y' = C_2 y \quad \Rightarrow \quad \dfrac{dy}{y} = C_2 dx
$$
Интегрируем и потенцируем
$$
\ln|y|=C_2 x + C_3 \quad \Rightarrow \quad \boxed{y = C_4 e^{C_2 x}}
$$


Запишем все решения:
$$
\boxed{y = A e^{B x}, \quad A, B \in \mathbb{R}.}
$$
Следует отметить, что первое решение $y = C$ включено в выражение выше при $B = 0$. Проверим решение:

```python
from sympy import *
x = symbols('x')
A, B = symbols('A B')
y = Function('y')(x)
```

```python
sol = A*exp(B*x) # первое решение
prime = diff(sol, x)  
dprime = diff(sol, x, x)  
LHS = sol * dprime
RHS = prime**2
simplify(LHS - RHS)
```

**Пример 2.** $yy''={y'}^2-{y'}^3$

Вводим $y'=z$, $y''=z'z$.

Подставляем в уравнение: $yz'z=z^2-z^3$.

Сразу видно тривиальное решение, если взять $z = 0$, то получится тождество. Поэтому $\boxed{y = C}$ будет первым решением. Упрощаем вид уравнения, чтобы получилось слева $z'$, а справа всё остальное.
$$
yz'z=z^2-z^3 \Rightarrow z'=\dfrac{z-z^2}{y} \Rightarrow \dfrac{dz}{dy}=\dfrac{z-z^2}{y} \Rightarrow \dfrac{dz}{z-z^2}=\dfrac{dy}{y}
$$

Как видно выше, переменные разделились. Раскладываем через простые дроби, чтобы затем проинтегрировать.
$$
\dfrac{1}{z-z^2} = \dfrac{1}{z(1-z)} = \frac{1}{z} + \frac{1}{1-z} \Rightarrow \left(\frac{1}{z} + \frac{1}{1-z}\right)dz = \frac{dy}{y}$$
Интегрируем
$$
\int \left(\frac{1}{z} + \frac{1}{1-z}\right) dz = \int \frac{dy}{y} \quad\Rightarrow\quad \ln\left|\frac{z}{1-z}\right| = \ln|y| + C
$$
Надо выразить $z$, для этого потенцируем.
$$
\left|\frac{z}{1-z}\right| = e^{C} |y|.
$$
Здесь сразу возникает вопрос, что делать с константой $C$, которая является произвольной постоянной интегрирования. Константа $e^{C}$ положительна всегда, потому что экспонента вещественного числа всегда положительна. 

Здесь можно пойти по такому пути. Ввести константу $K=e^C$. Тогда можно убрать модуль учитывая знак:
$$
\dfrac{z}{1-z} = \pm Ky
$$
Поскольку $\pm K$ ненулевая вещественная константа, можно её обозначить за $C_1 \neq 0$, если строго $C_1 \in \mathbb{R} \setminus \{0\}$. При $C_1$ получается частный случай $z = 0$. Он уже рассмотрен как решение.

Теперь выражаем $z$:
$$
z = C_1 y (1 - z) \Rightarrow z = C_1 y - C_1 y z \Rightarrow z + C_1 y z = C_1 y \Rightarrow z (1 + C_1 y) = C_1 y \Rightarrow z = \frac{C_1 y}{1 + C_1 y}
$$

Делаю обратную замену $z=y'$, через переворот (инвертирование):
$$
\frac{dy}{dx} = \frac{C_1 y}{1 + C_1 y} \Rightarrow \frac{dx}{dy} = \frac{1 + C_1 y}{C_1 y} \Rightarrow \frac{dx}{dy} = \frac{1}{C_1 y} + 1 \Rightarrow dx = \left(\frac{1}{C_1 y} + 1\right)dy
$$
Снова интегрируем 
$$
\int \left(\frac{1}{C_1 y} + 1\right) dy = \int dx \Rightarrow \dfrac{1}{C_1}\ln|y| + y = x + C_2
$$

Приведение к удобному виду, домножив обе части уравнения на $C_1 \neq 0$, как оговорено выше.
$$
\ln|y| + C_1 y = C_1 x + C_1 C_2
$$
Обозначим $C_1 C_2 = -C_3$ и перенесем $C_1 x$:
$$
\ln|y| + C_1 y = C_1 x + C_3
$$
Таким образом, интегрирование даёт неявное решение
$$
\boxed{\ln|y| + C_1 y = C_1 x + C_3}
$$

Получилось трансцендентное уравнение относительно $y$, одновременно присутствуют $\ln|⁡y|$ и $y$ в нелинейной комбинации. В общем случае нельзя выразить $y$ через элементарные функции от $x$ явно.

Таким образом, все решения:
$$
\boxed{y=C, \quad \ln|y| + C_1 y = C_1 x + C_3}, \quad C, C_1, C_3 \in \mathbb{R}.
$$

Проверить второе решение аналитически можно, но очень долго, надо формировать некоторый вид дифференцирования неявного уравнения.

Частично можно убедиться в правильности решения рассмотрев только одно частное решение при $C_1 = 1$, т.е. для функции вида:
$$
\ln|y| + y = x + C_3
$$
Дифференцируем неявно по $x$:
$$
\frac{1}{y} y' + y' = 1
$$
$$
y'\left(\frac{1}{y} + 1\right) = 1
$$
$$
y' = \frac{1}{\dfrac{1}{y} + 1} = \dfrac{y}{1 + y}, \tag{1}
$$

Находим $y''$, дифференцируя (1):
$$
y'' = \frac{y'(1+y) - y y'}{(1+y)^2}
      = \frac{y'(1+y - y)}{(1+y)^2}
      = \frac{y'}{(1+y)^2}
, \tag{2}
$$

Подставляем $y'$ из (1) в (2):
$$
y'' = \dfrac{\dfrac{y}{1+y}}{(1+y)^2} = \dfrac{y}{(1+y)^3}
$$

Проверка уравнения $y y'' = (y')^2 - (y')^3$

Левая часть:
$$
y y'' = y \frac{y}{(1+y)^3} = \frac{y^2}{(1+y)^3}
$$

Правая часть:
$$
(y')^2 - (y')^3 = \left(\frac{y}{1+y}\right)^2 - \left(\frac{y}{1+y}\right)^3
                = \frac{y^2}{(1+y)^2} - \frac{y^3}{(1+y)^3}
$$
$$
= \frac{y^2(1+y) - y^3}{(1+y)^3}
= \frac{y^2 + y^3 - y^3}{(1+y)^3}
= \frac{y^2}{(1+y)^3}
$$

В результате
$$
\frac{y^2}{(1+y)^3}
$$

Следовательно, при $C_1 = 1$ функция $y$, заданная неявно уравнением
$$
\ln|y| + y = x + C_3,
$$
удовлетворяет дифференциальному уравнению
$$
y y'' = (y')^2 - (y')^3.
$$

**Пример 3.** $y^{2}y''=y{y'}^2-2y'$

Вводим $y'=z$, $y''=z'z$.

Подставляем в уравнение: $y^{2}z'z=yz^2-2z$.

1. Первый вариант, если $z = 0$, т.е. $y'=0$. Тогда $y=C$ решение.

   Проверка: $C^2 \cdot 0 = C \cdot 0^2 - 2 \cdot 0$. Тождество есть, т.е. $\boxed{y=C}$ является решением. Это первый ответ.

2. Второй вариант, если $z \neq 0$. Пробуем решать $y^{2}z'z=yz^2-2z$. Принцип решения прост: пытаемся слева уравнения оставить $z'$, т.е. добраться до уравнения вида $z' = F(y,z)$ и посмотреть на его вид, т.е. классифицировать. Поэтому делим левую и правую части уравнения на $y^{2}z$. Получаем:
$$
\dfrac{y^2 z'z}{y^2z}=\dfrac{yz^2}{y^2z}-\dfrac{2z}{y^2z} \quad \Rightarrow \quad z'=\dfrac{z}{y}-\dfrac{2}{y^2}
$$
Это линейное неоднородное первого порядка. Запишем в более привычном виде:
$$
z'-\dfrac{1}{y}z=-\dfrac{2}{y^2}
$$
Такого типа уравнения были на второй по счету контрольной работе. Важно только, что $z=z(y)$. Любым известным способом добираемся до решения этого уравнения.
$$
z=Cy + \dfrac{1}{y}
$$
Чтобы добраться до $y(x)$ необходимо интегрировать левую и правую часть. Вот только здесь возможны два случая, когда $C = 0$ и когда $C \neq 0$.
    - $C=0 \Rightarrow y'=0 \cdot y + \dfrac{1}{y} \Rightarrow \dfrac{dy}{dx}=\dfrac{1}{y} \Rightarrow \boxed{y^2 = 2x + C}$
    - $C \neq 0 \Rightarrow y'=Cy + \dfrac{1}{y} \Rightarrow y'=\dfrac{Cy^2 + 1}{y} \Rightarrow \dfrac{dy}{dx}=\dfrac{Cy^2 + 1}{y}\Rightarrow \dfrac{ydy}{Cy^2 + 1}=dx$
   
    После интегрирования получаем: $\boxed{Cy^2=e^{2C(C_1+x)}-1}$

Запишем все решения для начального ДУ:
$$
\boxed{
y = C, \quad y^2 = 2x + C, \quad Cy^2=e^{2C(C_1+x)}-1, \quad C, C_1 \in \mathbb{R}.
}
$$

**Пример 4.** $x^{2}y''={y'}^2$

Вводим $y'=z$, $y''=z'$. Обратите внимание, $y''$ и $y'$ явно зависят от $x$, где $x$ является независимой переменной. Делать замену $y''=z'z$ некорректно, поскольку это смешает переменные $x$ и $y$ неразрешимым образом. 

Проверим, что получилось бы, если ошибочно взять $y''=z'z$, считая $z = z(y)$
$$
y'' = \dfrac{dz}{dx} = \dfrac{dz}{dy} \dfrac{dy}{dx} = z'z
$$
Подставляем в уравнение: $x^{2}z'z=z^2$ и получаем проблему, что $z$ зависит от $y$, а $x$ независимо присутствует в уравнении. Поэтому такой подход здесь неприменим.

Подставляем $y'=z$, $y''=z'$, считая $z=z(x)$, тогда $y''=\dfrac{dz}{dx}$. В случае явного вхождения $x$ в уравнение используем такой способ.

Подставим в уравнение:
$$
x^{2}\dfrac{dz}{dx}=z^2 \quad \Rightarrow \quad \dfrac{dz}{z^2} = \dfrac{dx}{x^2}
$$
Интегрируем, делаем обратную замену, разделяем переменные:
$$
-\dfrac{1}{z}=-\dfrac{1}{x} + C_1 \quad \Rightarrow \quad y'=\dfrac{x}{1-C_1 x} \quad \Rightarrow \quad dy=\dfrac{x dx}{1-C_1 x}
$$
Интегрируем:
$$
-\dfrac{1}{z}=-\dfrac{1}{x} + C_1 \quad \Rightarrow \quad y'=\dfrac{x}{1-C_1 x} \quad \Rightarrow \quad dy=\dfrac{x dx}{1-C_1 x}
$$

Здесь возникает два случая, $C_1 \neq 0$ и $C_1 = 0$. В первом случае ещё раз интегрируем:
$$
\int \frac{x}{1 - C_1 x} \, dx
$$
Пусть $u = 1 - C_1 x$, тогда $du = -C_1 dx$ $\Rightarrow$ $dx = -\dfrac{du}{C_1}, \quad x = \dfrac{1-u}{C_1}$.

$$
\begin{aligned}
\int \frac{x}{1 - C_1 x} \, dx 
&= \int \frac{\frac{1-u}{C_1}}{u} \left( -\frac{du}{C_1} \right) \\
&= -\frac{1}{C_1^2} \int \frac{1-u}{u} \, du \\
&= -\frac{1}{C_1^2} \int \left( \frac{1}{u} - 1 \right) du \\
&= -\frac{1}{C_1^2} \left( \ln|u| - u \right) + C_2 \\
&= -\frac{1}{C_1^2} \ln|1 - C_1 x| + \frac{1 - C_1 x}{C_1^2} + C_2 \\
&= -\frac{1}{C_1^2} \ln|1 - C_1 x| - \frac{x}{C_1} + \frac{1}{C_1^2} + C_2.
\end{aligned}
$$

Объединяем константы:
$$
\int \dfrac{x}{1 - C_1 x} \, dx = -\dfrac{x}{C_1} - \dfrac{1}{C_1^2} \ln|1 - C_1 x| + \widetilde{C}
$$

Во втором случае, $C_1 = 0$, т.е. интегрируем $y' = x \Rightarrow \boxed{y = \dfrac{x^2}{2} + C_2}$.

Запишем все решения для начального ДУ:
$$
\boxed{
y(x) = 
\begin{cases}
-\dfrac{x}{C_1} - \dfrac{1}{C_1^2} \ln|1 - C_1 x| + \widetilde{C}, & C_1 \neq 0, \\
\dfrac{x^2}{2} + C_2, \\
C,
\end{cases}
}
$$
где $C, C_1, C_2, \widetilde{C} \in \mathbb{R}$.

Проверим решения:

```python
import sympy as sp
import math
x = symbols('x')
A, B = symbols('A B')
y = Function('y')(x)
```

```python
sol = -x/A - 1/(A**2) * sp.log(1 - A*x) + B # первое решение
prime = diff(sol, x)  
dprime = diff(sol, x, x)  

LHS = x**2 * dprime
RHS = prime**2
simplify(LHS - RHS)
```

```python
sol = x**2/2 + A # второе решение
prime = diff(sol, x)  
dprime = diff(sol, x, x)  

LHS = x**2 * dprime
RHS = prime**2
simplify(LHS - RHS)
```

```python
sol = A # третье решение
prime = diff(sol, x)  
dprime = diff(sol, x, x)  

LHS = x**2 * dprime
RHS = prime**2
simplify(LHS - RHS)
```

**Пример 5.** $y''(e^x+1)+y'=0$

Вводим $y'=z$, $y''=z'$. Как в примере выше, $z = z(x)$.

Подставляем в уравнение:
$$
z'(e^x+1)+z=0 \quad \Rightarrow \quad z' + \dfrac{1}{e^x + 1} z=0
$$
Разделим переменные:
$$
\frac{dz}{z} = -\frac{dx}{e^x+1} \quad \Rightarrow \quad \ln|z| = -\int \frac{dx}{e^x+1}, \tag{1}
$$

Интегрируем, умножив числитель и знаменатель на $e^{-x}$:
$$
\int \frac{dx}{e^x+1} = \int \frac{e^{-x}}{1 + e^{-x}} dx
$$
Замена $t = 1 + e^{-x}$, $dt = -e^{-x} dx$ приводит к 
$$
\int \dfrac{e^{-x}}{1 + e^{-x}} dx = -\int \dfrac{dt}{t} = -\ln|t| + C = -\ln(1 + e^{-x}) + C
$$
Упрощаем 
$$
-\ln(1 + e^{-x}) + C = -\ln\left( \frac{e^x + 1}{e^x} \right) + C = -\ln(e^x + 1) + \ln(e^x) + C = x - \ln(e^x + 1) + C
$$
В (1) знак минус перед интегралом, не забываем его и потенцируем
$$
\ln|z| = \ln(e^x + 1) - x + C \quad \Rightarrow \quad \displaystyle e^{\ln|z|} = e^{\ln(e^x + 1) - x + C} \quad \Rightarrow \quad z = e^{\ln(e^x + 1)} e^{-x} e^{C}
$$
$$
z = C_1 (e^x + 1) e^{-x} \quad \Rightarrow \quad z = C_1 (1 + e^{-x})
$$
Интегрируем и получаем решение
$$
\dfrac{dy}{dx} = C_1 (1 + e^{-x}) \quad \Rightarrow \quad y = \int C_1 (1 + e^{-x}) dx \quad \Rightarrow \quad y = \boxed{C_1 (x - e^{-x}) + C_2}
$$
где $C_1, C_2 \in \mathbb{R}$.

**Пример 6.** ${y'}^2+2yy''=0$

Вводим $y'=z$, $y''=z'z$.

Подставляем в уравнение: $z^2+2yz'z=0$.

1. Первый вариант, если $z=0$, тогда $\boxed{y=C}$ является тривиальным решением.
2. Второй вариант, если $z \neq 0$. Делим уравнение на $z$:
$$
\dfrac{z^2}{z}+\dfrac{2yz'z}{z}=0 \quad \Rightarrow \quad z + 2yz' = 0
$$
Возможно разделение переменных, поэтому
$$
z+2y\dfrac{dz}{dy}=0 \quad \Rightarrow \quad \dfrac{dz}{z} = -\dfrac{dy}{2y}
$$
Интегрируем:
$$
\ln|z|=-\dfrac{1}{2}\ln|y|+C_1 \quad \Rightarrow \quad z = C y^{-1/2},
$$
где $C \neq 0$.
После этого разделяем переменные и интегрируем:
$$
\dfrac{dy}{dx}=C y^{-1/2} \quad \Rightarrow \quad y^{1/2} dy = C dx \quad \Rightarrow \quad \dfrac{2}{3} y^{3/2} = C x + C_2,
$$
Возводим в квадрат, переобозначая константы и получая второе решение:
$$
y^{3/2} = C x + C_2 \quad \Rightarrow \quad \boxed{y^3 = C_1 (x + C_2)^2}.
$$
Запишем все решения для начального ДУ:
$$
\boxed{
y = C, \quad y^3 = C_1 (x + C_2)^2, \quad C, C_1, C_2 \in \mathbb{R}.
}
$$

Проверим решения:
